# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x = 'Gender',data = df)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen =  df.groupby(['Gender'])['Amount'].sum().reset_index()
sales_gen
```

```python

df1=sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
plt.title('Total amount By Gender')
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'])['Amount'].sum().reset_index().sort_values(ascending=False,by='Amount')
plt.title('Total amount By Age Group')
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby('State')['Orders'].sum().reset_index().sort_values(ascending=False,by='Orders').head(10)
sns.barplot(data = sales_state, y = 'State',x= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby('State')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
plt.title('Total amount By States')
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, y = 'State',x= 'Amount')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(15,12)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
plt.title('Total amount By Martial Status')
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)
plt.title('Total amount By Occupation')
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, y = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
plt.title('Total amount By Product_Category')
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, y = 'Product_Category',x= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
Product = df.groupby(['Product_Category'])['Orders'].sum().reset_index().sort_values(by='Orders', ascending=False).head(10)
Product
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = Product,  x='Product_Category',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)
Product_Orderd = df.groupby(['Product_ID'])['Orders'].sum().reset_index().sort_values(by='Orders', ascending=False).head(10)
Product_Orderd
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = Product_Orderd,  x='Product_ID',y= 'Orders')
```

## Conclusion:

*Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category .*
